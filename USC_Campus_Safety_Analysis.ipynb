```python
import seaborn as sns
import matplotlib.pyplot as plt

# Counting occurrences of each crime type
crime_counts = usc_area['Crm Cd Desc'].value_counts()

# Plotting
plt.figure(figsize=(50, 20))
sns.barplot(x=crime_counts.values, y=crime_counts.index, palette='viridis')
plt.title('Frequency of Different Types of Crimes Around USC')
plt.xlabel('Number of Incidents')
plt.ylabel('Crime Type')
plt.show()
```

```python
# Convert date column to datetime format if not already
usc_area['Date Rptd'] = pd.to_datetime(usc_area['Date Rptd'])

# Plotting
plt.figure(figsize=(10, 5))
usc_area['Date Rptd'].dt.date.hist(bins=30, color='skyblue')
plt.title('Crime Occurrences Over Time Near USC')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.show()
```
